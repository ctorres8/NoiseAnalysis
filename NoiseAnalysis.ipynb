# Análisis estadístico de Ruidos Reales

Para el análisis se utilizaron las base de datos del MIT (MIT-BIH Noise Stress Test Database). Se analizaron los ruidos realistas correspondiende a:
* Corrimiento de la línea de Base (BW)
* Movimiento de los Electrodos (EM)
* Dispostivos Motrices (MA)


```python
"""
** Baseline Wander (BW) **

Source: record nstdb/bw
val has 2 rows (signals) and 650000 columns (samples/signal)
Duration:    30:05
Sampling frequency: 360 Hz  Sampling interval: 0.002777777778 sec
Row   Signal  Gain   Base   Units
1     noise1   200    0      mV
2     noise2   200    0      mV

------------------------------------------

** Electrode Movement (EM) **

Source: record nstdb/em
val has 2 rows (signals) and 650000 columns (samples/signal)
Duration:    30:05
Sampling frequency: 360 Hz  Sampling interval: 0.002777777778 sec
Row   Signal  Gain    Base    Units
1     noise1  200      0       mV
2     noise2  200      0       mV

------------------------------------------

** Motion Artifacts (MA) **

Source: record nstdb/ma
val has 2 rows (signals) and 650000 columns (samples/signal)
Duration:    30:05
Sampling frequency: 360 Hz  Sampling interval: 0.002777777778 sec
Row   Signal    Gain    Base  Units
1     noise1    200      0     mV
2     noise2    200      0     mV

"""

import numpy as np
from scipy import signal as sig
import matplotlib.pyplot as plt
import scipy.io as sio
import pandas as pd
from IPython.display import HTML
#import os
```

```python
bw = sio.loadmat('bwm.mat')['val']
em = sio.loadmat('emm.mat')['val']
ma = sio.loadmat('mam.mat')['val']


fs = 360#Hz
```

```python
#Media
bw_mean = [bw[0].mean(),bw[1].mean()]
em_mean = [em[0].mean(),em[1].mean()]
ma_mean = [ma[0].mean(),ma[1].mean()]

media =[bw_mean[0],bw_mean[1],em_mean[0],em_mean[1],ma_mean[0],ma_mean[1]]

#Media Absoluta
bw_mem = [abs(bw[0]).mean(),abs(bw[1]).mean()]
em_mem = [abs(em[0]).mean(),abs(em[1]).mean()]
ma_mem = [abs(ma[0]).mean(),abs(ma[1]).mean()]

mem = [bw_mem[0],bw_mem[1],em_mem[0],em_mem[1],ma_mem[0],ma_mem[1]]

# Desviacion Standard
bw_std = [bw[0].std(),bw[1].std()]
em_std = [em[0].std(),em[1].std()]
ma_std = [ma[0].std(),ma[1].std()]

std = [bw_std[0],bw_std[1],em_std[0],em_std[1],ma_std[0],ma_std[1]]

#Varianza
bw_var = [bw[0].var(),bw[1].var()]
em_var = [em[0].var(),em[1].var()]
ma_var = [ma[0].var(),ma[1].var()]

var = [bw_var[0],bw_var[1],em_var[0],em_var[1],ma_var[0],ma_var[1]]
```

```python
filas =['BW0','BW1','EM0','EM1','MA0','MA1']

df = pd.DataFrame({'Media':media,
                   'MeM':mem,
                   'STD':std,
                   'VAR':var
                   },index=filas)

HTML(df.to_html())
```

## Análisis de la DEP

```python
#Ventana tomada

t1=80000
t2=200000
win = np.arange(t1,t2+1,1)/fs

bw_w = bw[0][t1:t2+1],bw[1][t1:t2+1]
em_w = em[0][t1:t2+1],em[1][t1:t2+1]
ma_w = ma[0][t1:t2+1],ma[1][t1:t2+1]

n_win = len(bw_w)
```

### Ventana tomada para el ruido BW

```python
plt.figure(figsize=[12,8])
for j in range(1,3):
    plt.subplot(2,1,j)
    #plt.tight_layout(2)
    if j==1:
        plt.plot(win,bw_w[j-1],label='BW0')
        #plt.title(" Ventana BW0 ")
        plt.xlabel("Tiempo [seg]")
        plt.ylabel("Tensión[V]")
        axes_bw0_w = plt.gca()
        axes_bw0_w.legend()
    if j==2:
        plt.plot(win,bw_w[j-1],label='BW1')
        #plt.title("Ventana BW1")
        plt.xlabel("Tiempo [seg]")
        plt.ylabel("Tensión[V]")
        axes_bw1_w = plt.gca()
        axes_bw1_w.legend()
        
    plt.grid(True)
```

### Ventanas tomadas para el ruido EM

```python
plt.figure("Ventanas tomadas", figsize=[12,8])
for j in range(1,3):
    plt.subplot(2,1,j)
    #plt.tight_layout(2)
    if j==1:
        plt.plot(win,em_w[j-1],label='EM0')
        #plt.title(" Ventana BW0 ")
        plt.xlabel("Tiempo [seg]")
        plt.ylabel("Tensión[V]")
        axes_em0_w = plt.gca()
        axes_em0_w.legend()
    if j==2:
        plt.plot(win,em_w[j-1],label='EM1')
        #plt.title("Ventana BW1")
        plt.xlabel("Tiempo [seg]")
        plt.ylabel("Tensión[V]")
        axes_em1_w = plt.gca()
        axes_em1_w.legend()
        
    plt.grid(True)
```

### Ventanas tomadas para el ruido MA

```python
plt.figure("Ventanas tomadas", figsize=[12,8])
for j in range(1,3):
    plt.subplot(2,1,j)
    #plt.tight_layout(2)
    if j==1:
        plt.plot(win,ma_w[j-1],label='MA0')
        #plt.title(" Ventana BW0 ")
        plt.xlabel("Tiempo [seg]")
        plt.ylabel("Tensión[V]")
        axes_ma0_w = plt.gca()
        axes_ma0_w.legend()
    if j==2:
        plt.plot(win,ma_w[j-1],label='MA1')
        #plt.title("Ventana BW1")
        plt.xlabel("Tiempo [seg]")
        plt.ylabel("Tensión[V]")
        axes_ma1_w = plt.gca()
        axes_ma1_w.legend()
        
    plt.grid(True)
```

# DEP

```python
#Periodograma
n=14

Fxx_bw0,Pxx_bw0=sig.welch(bw_w[0],fs,window='hanning',nperseg=2**n,noverlap=(2**n)/4, nfft=2*(2**n))
Fxx_bw1,Pxx_bw1=sig.welch(bw_w[1],fs,window='hanning',nperseg=2**n,noverlap=(2**n)/4, nfft=2*(2**n))

Fxx_em0,Pxx_em0=sig.welch(em_w[0],fs,window='hanning',nperseg=2**n,noverlap=(2**n)/4, nfft=2*(2**n))
Fxx_em1,Pxx_em1=sig.welch(em_w[1],fs,window='hanning',nperseg=2**n,noverlap=(2**n)/4, nfft=2*(2**n))

Fxx_ma0,Pxx_ma0=sig.welch(ma_w[0],fs,window='hanning',nperseg=2**n,noverlap=(2**n)/4, nfft=2*(2**n))
Fxx_ma1,Pxx_ma1=sig.welch(ma_w[1],fs,window='hanning',nperseg=2**n,noverlap=(2**n)/4, nfft=2*(2**n))
```

```python
plt.figure("Densidad de potencia BW Noise",figsize=[10,8])

plt.plot(Fxx_bw0,Pxx_bw0,label ='Potencia BW0')
plt.plot(Fxx_bw1,Pxx_bw1,label ='Potencia BW1')

plt.xlim(0,2,0.1)
plt.title('DEP de BW')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud ')
plt.grid(which='both', axis='both')

axes_dep_bw = plt.gca()
axes_dep_bw.legend()
```

```python
plt.figure("Densidad de potencia EM Noise",figsize=[10,8])

plt.plot(Fxx_em0,Pxx_em0,label ='Potencia EM0')
plt.plot(Fxx_em1,Pxx_em1,label ='Potencia EM1')

plt.xlim(0,10,0.1)
plt.title('DEP de EM')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud ')
plt.grid(which='both', axis='both')

axes_dep_em = plt.gca()
axes_dep_em.legend()
```

```python
plt.figure("Densidad de potencia MA Noise",figsize=[10,8])

plt.plot(Fxx_ma0,Pxx_ma0,label ='Potencia MA0')
plt.plot(Fxx_ma1,Pxx_ma1,label ='Potencia MA1')

plt.xlim(0,3,0.1)
plt.title('DEP de MA')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud ')
plt.grid(which='both', axis='both')

axes_dep_ma = plt.gca()
axes_dep_ma.legend()
```
